```python
text = "Dungeons and dragons is a cooperative storytelling game in which players assume the roles of heroes, or Player Characters (PCs), undergoing an adventure. One person, acting as the Dungeon Master (DM), is responsible for controlling the actions of all the characters the PCs encounter, either good or bad. The DM is the only player with previous knowledge of events the players encounter and thus, is responsible for revealing the story as it progresses. The players decide how their characters will react to different situations and attempt to perform actions such as attacking monsters, crafting weapons and armor, or interacting with Non Player Characters (NPCs) by rolling dice and adding bonuses they receive from items or innate abilities which are decided upon creation of the character. There are extensive rules in place to account for most situations ranging from bartering for cabbages from a lowly farmer to battling wizards, dragons, and in some cases extra-planar beings or demigods. The DM may strictly adhere to these rules or may omit them if it would make more sense. The biggest benefit of playing a table top RPG like Dungeons and Dragons is that it can be adapted to put the players in the story they want to be in. It is literally only limited to the imagination of the people at the table."
tokens = text.encode("utf-8")
tokens = list(map(int, tokens))

print("---")
print(tokens)
print("---")
print("length:", len(tokens))
```

```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]): 
        """
        ids -> [68, 117, 110, 85]
        ids[1:] ->  117, 110, 85  
        zip pairs them element wise:
            (68, 117)
            (117, 110)
            (110, 85)    
        """
        if pair in counts: # if pair is not in count, start at 0, otherwise take existing count and add one.
            counts[pair] += 1
        else:
            counts[pair] = 1 
    return counts

stats = get_stats(tokens)
print(sorted(((v,k) for k,v in stats.items()), reverse = True))
```

```python
top_pair = max(stats, key=stats.get) # rank the value
top_pair
```

```python
def merge(ids, pair, idx):
    newids = []
    i = 0
    while i < len(ids):
        if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            newids.append(idx)
            i +=2
        else:
            newids.append(ids[i])
            i += 1
    return newids

print(merge([5, 6, 6, 7, 9, 1], (6, 7), 99))

tokens2 = merge(tokens, top_pair, 256)
print(tokens2)
print(len(tokens2))
```

```python
vocab_size = 276
num_merges = vocab_size - 256
ids = list(tokens)

merges = {}

for i in range(num_merges):
    stats = get_stats(ids)
    pair = max(stats, key=stats.get)
    idx = 256 + i
    print(f"merging {pair} in to new token {idx}")
    ids = merge(ids, pair, idx)
    merges[pair] = idx
```
